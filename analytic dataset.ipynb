```python
from music21 import *
import pandas as pd
import os
```

```python
csvFiles=[]
for f in os.listdir('./csv_files'):
    if f.find('.csv')>0:
        csvFiles.append(f)
```

```python
csvFiles
```

```python
i=0
while (i<len(csvFiles)):
    #print(i)
    #print(csvFiles[i])
    df_piece=pd.read_csv('csv_files/'+csvFiles[i])
    if i==0:
        df=df_piece
    else:
        df=pd.concat([df,df_piece],axis=0,sort=True)
    i+=1
```

```python
df
```

```python
workList=pd.read_csv('workList.csv')
```

```python
workList.columns
```

```python
df_pieceInfo=pd.merge(df,workList,how='outer',on='Composer')
```
